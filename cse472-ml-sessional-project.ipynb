## MultiResUNet

- [Paper](https://doi.org/10.1016/j.neunet.2019.08.025)
- [Codebase](https://github.com/nibtehaz/MultiResUNet/tree/master)

```python
def conv2d_bn(x, filters, num_row, num_col, padding='same', strides=(1, 1), activation='relu', name=None):
    '''
    2D Convolutional layers
    
    Arguments:
        x {keras layer} -- input layer 
        filters {int} -- number of filters
        num_row {int} -- number of rows in filters
        num_col {int} -- number of columns in filters
    
    Keyword Arguments:
        padding {str} -- mode of padding (default: {'same'})
        strides {tuple} -- stride of convolution operation (default: {(1, 1)})
        activation {str} -- activation function (default: {'relu'})
        name {str} -- name of the layer (default: {None})
    
    Returns:
        [keras layer] -- [output layer]
    '''

    x = Conv2D(filters, (num_row, num_col), strides=strides, padding=padding, use_bias=False)(x)
    x = BatchNormalization(axis=3, scale=False)(x)

    if(activation == None):
        return x

    x = Activation(activation, name=name)(x)

    return x


def trans_conv2d_bn(x, filters, num_row, num_col, padding='same', strides=(2, 2), name=None):
    '''
    2D Transposed Convolutional layers
    
    Arguments:
        x {keras layer} -- input layer 
        filters {int} -- number of filters
        num_row {int} -- number of rows in filters
        num_col {int} -- number of columns in filters
    
    Keyword Arguments:
        padding {str} -- mode of padding (default: {'same'})
        strides {tuple} -- stride of convolution operation (default: {(2, 2)})
        name {str} -- name of the layer (default: {None})
    
    Returns:
        [keras layer] -- [output layer]
    '''

    x = Conv2DTranspose(filters, (num_row, num_col), strides=strides, padding=padding)(x)
    x = BatchNormalization(axis=3, scale=False)(x)
    
    return x


def MultiResBlock(U, inp, alpha = 1.67):
    '''
    MultiRes Block
    
    Arguments:
        U {int} -- Number of filters in a corrsponding UNet stage
        inp {keras layer} -- input layer 
    
    Returns:
        [keras layer] -- [output layer]
    '''

    W = alpha * U

    shortcut = inp

    shortcut = conv2d_bn(shortcut, int(W*0.167) + int(W*0.333) +
                         int(W*0.5), 1, 1, activation=None, padding='same')

    conv3x3 = conv2d_bn(inp, int(W*0.167), 3, 3,
                        activation='relu', padding='same')

    conv5x5 = conv2d_bn(conv3x3, int(W*0.333), 3, 3,
                        activation='relu', padding='same')

    conv7x7 = conv2d_bn(conv5x5, int(W*0.5), 3, 3,
                        activation='relu', padding='same')

    out = concatenate([conv3x3, conv5x5, conv7x7], axis=3)
    out = BatchNormalization(axis=3)(out)

    out = add([shortcut, out])
    out = Activation('relu')(out)
    out = BatchNormalization(axis=3)(out)

    return out


def ResPath(filters, length, inp):
    '''
    ResPath
    
    Arguments:
        filters {int} -- [description]
        length {int} -- length of ResPath
        inp {keras layer} -- input layer 
    
    Returns:
        [keras layer] -- [output layer]
    '''


    shortcut = inp
    shortcut = conv2d_bn(shortcut, filters, 1, 1,
                         activation=None, padding='same')

    out = conv2d_bn(inp, filters, 3, 3, activation='relu', padding='same')

    out = add([shortcut, out])
    out = Activation('relu')(out)
    out = BatchNormalization(axis=3)(out)

    for i in range(length-1):

        shortcut = out
        shortcut = conv2d_bn(shortcut, filters, 1, 1,
                             activation=None, padding='same')

        out = conv2d_bn(out, filters, 3, 3, activation='relu', padding='same')

        out = add([shortcut, out])
        out = Activation('relu')(out)
        out = BatchNormalization(axis=3)(out)

    return out


def MultiResUnet(height, width, n_channels):
    '''
    MultiResUNet
    
    Arguments:
        height {int} -- height of image 
        width {int} -- width of image 
        n_channels {int} -- number of channels in image
    
    Returns:
        [keras model] -- MultiResUNet model
    '''


    inputs = Input((height, width, n_channels))

    mresblock1 = MultiResBlock(32, inputs)
    pool1 = MaxPooling2D(pool_size=(2, 2))(mresblock1)
    mresblock1 = ResPath(32, 4, mresblock1)

    mresblock2 = MultiResBlock(32*2, pool1)
    pool2 = MaxPooling2D(pool_size=(2, 2))(mresblock2)
    mresblock2 = ResPath(32*2, 3, mresblock2)

    mresblock3 = MultiResBlock(32*4, pool2)
    pool3 = MaxPooling2D(pool_size=(2, 2))(mresblock3)
    mresblock3 = ResPath(32*4, 2, mresblock3)

    mresblock4 = MultiResBlock(32*8, pool3)
    pool4 = MaxPooling2D(pool_size=(2, 2))(mresblock4)
    mresblock4 = ResPath(32*8, 1, mresblock4)

    mresblock5 = MultiResBlock(32*16, pool4)

    up6 = concatenate([Conv2DTranspose(
        32*8, (2, 2), strides=(2, 2), padding='same')(mresblock5), mresblock4], axis=3)
    mresblock6 = MultiResBlock(32*8, up6)

    up7 = concatenate([Conv2DTranspose(
        32*4, (2, 2), strides=(2, 2), padding='same')(mresblock6), mresblock3], axis=3)
    mresblock7 = MultiResBlock(32*4, up7)

    up8 = concatenate([Conv2DTranspose(
        32*2, (2, 2), strides=(2, 2), padding='same')(mresblock7), mresblock2], axis=3)
    mresblock8 = MultiResBlock(32*2, up8)

    up9 = concatenate([Conv2DTranspose(32, (2, 2), strides=(
        2, 2), padding='same')(mresblock8), mresblock1], axis=3)
    mresblock9 = MultiResBlock(32, up9)

    conv10 = conv2d_bn(mresblock9, 1, 1, 1, activation='sigmoid')
    
    model = Model(inputs=[inputs], outputs=[conv10])

    return model
```

## Modified U-Net

- [Paper](https://pubmed.ncbi.nlm.nih.gov/35304675/)

```python
def conv_block(input_tensor, filters, kernel_sizes=[1, 3, 5]):
    """Convolutional block with filters of multiple kernel sizes."""
    convs = []
    for k in kernel_sizes:
        conv = layers.Conv2D(filters, (k, k), padding='same', activation='relu')(input_tensor)
        convs.append(conv)
    merged = layers.Concatenate()(convs)
    bn = layers.BatchNormalization()(merged)
    return bn

def modified_unet(input_shape=(128, 128, 3)):
    """Builds the modified U-Net model as described in the paper."""
    inputs = layers.Input(shape=input_shape)

    # Contracting Path
    c1 = conv_block(inputs, 64)
    p1 = layers.MaxPooling2D((2, 2))(c1)
    p1 = layers.Dropout(0.1)(p1)

    c2 = conv_block(p1, 128)
    p2 = layers.MaxPooling2D((2, 2))(c2)
    p2 = layers.Dropout(0.1)(p2)

    c3 = conv_block(p2, 256)
    p3 = layers.MaxPooling2D((2, 2))(c3)
    p3 = layers.Dropout(0.1)(p3)

    c4 = conv_block(p3, 512)
    p4 = layers.MaxPooling2D((2, 2))(c4)
    p4 = layers.Dropout(0.1)(p4)

    # Bottleneck
    bn = conv_block(p4, 1024)

    # Expansive Path
    u4 = layers.Conv2DTranspose(512, (2, 2), strides=(2, 2), padding='same')(bn)
    u4 = layers.Concatenate()([u4, c4])
    u4 = layers.Dropout(0.1)(u4)
    c5 = conv_block(u4, 512)

    u3 = layers.Conv2DTranspose(256, (2, 2), strides=(2, 2), padding='same')(c5)
    u3 = layers.Concatenate()([u3, c3])
    u3 = layers.Dropout(0.1)(u3)
    c6 = conv_block(u3, 256)

    u2 = layers.Conv2DTranspose(128, (2, 2), strides=(2, 2), padding='same')(c6)
    u2 = layers.Concatenate()([u2, c2])
    u2 = layers.Dropout(0.1)(u2)
    c7 = conv_block(u2, 128)

    u1 = layers.Conv2DTranspose(64, (2, 2), strides=(2, 2), padding='same')(c7)
    u1 = layers.Concatenate()([u1, c1])
    u1 = layers.Dropout(0.1)(u1)
    c8 = conv_block(u1, 64)

    # Output Layer
    outputs = layers.Conv2D(1, (1, 1), activation='sigmoid')(c8)

    model = models.Model(inputs, outputs)
    return model
```

## Our Model

Key improvements:
- Swish Activation Function 
- SpatialDropout2D
- Bilinear Interpolation
- Combined Loss Function: Binary Cross-Entropy + Dice Loss

```python
def our_conv_block(input_tensor, filters, kernel_sizes=[1, 3, 5], activation='swish', dropout_rate=0.1):
    """Convolutional block with filters of multiple kernel sizes and improved activation/regularization."""
    convs = []
    for k in kernel_sizes:
        conv = layers.Conv2D(filters, (k, k), padding='same', activation=activation)(input_tensor)
        conv = layers.BatchNormalization()(conv)
        convs.append(conv)
    merged = layers.Concatenate()(convs)
    merged = layers.SpatialDropout2D(dropout_rate)(merged)
    return merged

def our_modified_unet(input_shape=(128, 128, 3)):
    """Builds the modified U-Net model with updated activation, regularization, upsampling, and loss functions."""
    inputs = layers.Input(shape=input_shape)

    # Contracting Path
    c1 = our_conv_block(inputs, 64)
    p1 = layers.MaxPooling2D((2, 2))(c1)

    c2 = our_conv_block(p1, 128)
    p2 = layers.MaxPooling2D((2, 2))(c2)

    c3 = our_conv_block(p2, 256)
    p3 = layers.MaxPooling2D((2, 2))(c3)

    c4 = our_conv_block(p3, 512)
    p4 = layers.MaxPooling2D((2, 2))(c4)

    # Bottleneck
    bn = our_conv_block(p4, 1024)

    # Expansive Path
    u4 = layers.UpSampling2D((2, 2), interpolation='bilinear')(bn)
    u4 = layers.Concatenate()([u4, c4])
    c5 = our_conv_block(u4, 512)

    u3 = layers.UpSampling2D((2, 2), interpolation='bilinear')(c5)
    u3 = layers.Concatenate()([u3, c3])
    c6 = our_conv_block(u3, 256)

    u2 = layers.UpSampling2D((2, 2), interpolation='bilinear')(c6)
    u2 = layers.Concatenate()([u2, c2])
    c7 = our_conv_block(u2, 128)

    u1 = layers.UpSampling2D((2, 2), interpolation='bilinear')(c7)
    u1 = layers.Concatenate()([u1, c1])
    c8 = our_conv_block(u1, 64)

    # Output Layer
    outputs = layers.Conv2D(1, (1, 1), activation='sigmoid')(c8)

    model = models.Model(inputs, outputs)
    return model

def dice_loss(y_true, y_pred, smooth=1e-6):
    """Dice loss for binary segmentation."""
    y_true_f = tf.keras.backend.flatten(y_true)
    y_pred_f = tf.keras.backend.flatten(y_pred)
    intersection = tf.keras.backend.sum(y_true_f * y_pred_f)
    return 1 - (2. * intersection + smooth) / (tf.keras.backend.sum(y_true_f) + tf.keras.backend.sum(y_pred_f) + smooth)

def combined_loss(y_true, y_pred):
    """Combines Binary Cross-Entropy with Dice Loss."""
    bce = tf.keras.losses.binary_crossentropy(y_true, y_pred)
    d_loss = dice_loss(y_true, y_pred)
    return bce + d_loss
```

```python
def dice_coef(y_true, y_pred):
    smooth = 1e-6
    y_true_f = K.flatten(y_true)
    y_pred_f = K.flatten(y_pred)
    intersection = K.sum(y_true_f * y_pred_f)
    return (2. * intersection + smooth) / (K.sum(y_true_f) + K.sum(y_pred_f) + smooth)

def jacard(y_true, y_pred):
    smooth = 1e-6
    y_true_f = K.flatten(y_true)
    y_pred_f = K.flatten(y_pred)
    intersection = K.sum ( y_true_f * y_pred_f)
    union = K.sum ( y_true_f + y_pred_f - y_true_f * y_pred_f)

    return (intersection + smooth) / (union + smooth)

def saveModel(model, outfilepath='models'):

    model_json = model.to_json()

    try:
        os.makedirs(outfilepath)
    except:
        pass
    
    fp = open(f'{outfilepath}/modelP.json','w')
    fp.write(model_json)
    model.save_weights(f'{outfilepath}/modelW.weights.h5')

def evaluateModel(model, X_test, Y_test, batchSize, outfilepath='models', resfilepath='results', val=True):
    
    try:
        os.makedirs(resfilepath)
    except:
        pass 
    

    yp = model.predict(x=X_test, batch_size=batchSize, verbose=1)
    yp = np.round(yp,0)

    smooth = 1e-6
    
    jacard = 0
    dice = 0
    jacard_scores = []
    
    for i in range(len(Y_test)):
        yp_2 = yp[i].ravel()
        y2 = Y_test[i].ravel()
        
        intersection = yp_2 * y2
        union = yp_2 + y2 - intersection

        jacard_score = ((np.sum(intersection) + smooth) / (np.sum(union) + smooth))  
        jacard += jacard_score
        
        dice += (2. * np.sum(intersection) + smooth) / (np.sum(yp_2) + np.sum(y2) + smooth)

        if np.sum(yp_2) == 0 and np.sum(y2) == 0:
            continue
        jacard_scores.append((i, jacard_score))

    jacard /= len(Y_test)
    dice /= len(Y_test)
    
    print('Jacard Index : '+str(jacard))
    print('Dice Coefficient : '+str(dice))
    
    # Sort the scores by Jaccard index
    jacard_scores = sorted(jacard_scores, key=lambda x: x[1], reverse=True)
    top_bottom_10_scores = [item[0] for item in jacard_scores[:5]] + [item[0] for item in jacard_scores[-5:]]

    for idx in top_bottom_10_scores:
        # Prepare plots
        fig, axes = plt.subplots(1, 3, figsize=(12, 5))  # 3 columns: Input, Ground Truth, Prediction
        fig.suptitle("Testing Images: Input vs Ground Truth vs Prediction", fontsize=16)

        input_image = X_test[idx]  # Shape: (height, width, channels)
        ground_truth = Y_test[idx]  # Shape: (height, width, 1)
        prediction = yp[idx]
        
        # Plot input image
        axes[0].imshow(input_image.squeeze(), cmap="gray")
        axes[0].set_title("Input Image")
        axes[0].axis("off")

        # Plot ground truth
        axes[1].imshow(ground_truth.squeeze(), cmap="gray")
        axes[1].set_title("Ground Truth")
        axes[1].axis("off")

        # Plot prediction
        axes[2].imshow(prediction.squeeze(), cmap="gray")
        axes[2].set_title("Prediction")
        axes[2].axis("off")

        intersection = yp[idx].ravel() * Y_test[idx].ravel()
        union = yp[idx].ravel() + Y_test[idx].ravel() - intersection
        
        jacard_score = ((np.sum(intersection) + smooth) / (np.sum(union) + smooth))  
        plt.suptitle('Jacard Index: '+ str(np.sum(intersection)) +' / '+ str(np.sum(union)) +' = '+str(jacard_score))

        # Adjust layout
        # plt.tight_layout(rect=[0, 0, 1, 0.96])  # Leave space for the title

        # Adjust spacing to remove extra white space
        plt.subplots_adjust(wspace=0.5, hspace=0)  # Minimal horizontal and vertical spacing
        plt.tight_layout(pad=0.5)  # Remove padding around the figure
        
        # Save the figure without additional margins
        plt.savefig(resfilepath + '/' + str(i) +'.png',format='png', bbox_inches='tight', pad_inches=0.5)
        
        # plt.savefig()
        plt.close()


    if val:
        fp = open(f'{outfilepath}/log.txt','a')
        fp.write(str(jacard)+'\n')
        fp.close()
    
        fp = open(f'{outfilepath}/best.txt','r')
        best = fp.read()
        fp.close()
    
        if(jacard>float(best)):
            print('***********************************************')
            print('Jacard Index improved from '+str(best)+' to '+str(jacard))
            print('***********************************************')
            fp = open(f'{outfilepath}/best.txt','w')
            fp.write(str(jacard))
            fp.close()
    
            saveModel(model, outfilepath=outfilepath)


def trainStep(model, X_train, Y_train, X_test, Y_test, epochs, batchSize, outfilepath, resfilepath):

    
    for epoch in range(epochs):
        print('Epoch : {}'.format(epoch+1))
        model.fit(x=X_train, y=Y_train, batch_size=batchSize, epochs=1, verbose=1)     

        evaluateModel(model,X_test, Y_test,batchSize, outfilepath=outfilepath, resfilepath=resfilepath)

    return model 

# def combined_loss(y_true, y_pred):
#     smooth = 1e-6
#     # Dice loss
#     y_true_f = K.flatten(y_true)
#     y_pred_f = K.flatten(y_pred)
#     intersection = K.sum(y_true_f * y_pred_f)
#     dice_loss = 1 - (2. * intersection + smooth) / (K.sum(y_true_f) + K.sum(y_pred_f) + smooth)
#     # Binary cross-entropy
#     bce_loss = K.binary_crossentropy(y_true, y_pred)
#     return dice_loss + bce_loss
```
